```python
import wikipedia, re
print wikipedia.summary("List_of_sports")
```

```python
page = wikipedia.page("List_of_sports")
print page.content
```

```python
sportslist = re.split('\(|\n', page.content.encode('utf8'))
sportslist
```

```python
# get rid of the non sport names items, ending ')', names longer than 5 words

sportslist = [re.sub('\)','', x) for x in sportslist if x != '' and '=' not in x and len(x.split()) <=5 ]
sportslist
```

```python
f = open('sportslist.txt', 'w')
f.write("\n".join(sportslist))
f.close
```
