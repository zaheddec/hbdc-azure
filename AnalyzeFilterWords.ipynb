```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import nltk
from nltk.stem.porter import PorterStemmer
import re
import string
from nltk.corpus import stopwords


DF = pd.read_csv('FilteredTweets-sedentary.csv')
# get only the labelled tweets, in last 2 columns
DF = DF[DF['Sit == True  (1), else (0)'].notnull() & DF['Relevant == True (1), else (0)'].notnull()]

DF
```

```python
#Find common terms for each subset
def commonTerms(tweets): #df.text.tolist() here

	porter = nltk.PorterStemmer()
	all_tokens = []


	for i, value in enumerate(tweets):
	    
	    lowered = value.lower()
	    
	    tokens = nltk.word_tokenize(lowered)

	    # filter
	    tokens = [item for item in tokens if re.search(r'^[A-Za-z]+$', item)]

	    tokens = [item for item in tokens if item not in stopwords.words('english')]

	    stemmed_tokens = [porter.stem(item) for item in tokens]

	    all_tokens = all_tokens + stemmed_tokens



	freq = nltk.FreqDist(all_tokens)


	freq_terms = pd.DataFrame(freq.most_common(), columns =['words', 'frequency'])

	return freq_terms
```

```python
# stratify the sample

relevant = DF[DF['Relevant == True (1), else (0)'] == 1.0]
not_relevant = DF[DF['Relevant == True (1), else (0)'] == 0.0]

sit = DF[DF['Sit == True  (1), else (0)'] == 1.0]
not_sit = DF[DF['Sit == True  (1), else (0)'] == 0.0]
```
